```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv('APOT_Project_Data_Jan 2017_to_Sept_2022FinalVersion.csv', encoding='latin1')
data 
```

```python
ems_df = pd.DataFrame(data)
ems_df
```

```python
# rename ID columns b/c has space in name 
ems_df = ems_df.rename(columns={"ID ":"ID", "Date ": "Date", "Hospital Code":"Hospital_Code", "AgenyAndUnit": "AgencyAndUnit"})
```

```python
ems_df = ems_df.dropna()
```

```python
ems_df2 = pd.DataFrame(ems_df.reset_index(drop=True))
ems_df2
```

```python
# Convert 'Date' to datetime data type
dates = pd.to_datetime(ems_df2['Date'])
dates
```

```python
# Convert 'ArrivalDateTime' to datetime data type
arrival_time = pd.to_datetime(ems_df2['ArrivalDateTime'])
arrival_time
```

```python
# create merge_date column so can merge new datatype into new dataframe 
merge_date = ems_df2['Date']
ID = ems_df2['ID']
hospital_code = ems_df2['Hospital_Code']
lat = ems_df2['Lat']
long = ems_df2['Long']
apot = ems_df2['APOT']
impression = ems_df2['Impression']
postal_code = ems_df2['PostalCode']
agency = ems_df2['AgencyNumber']
unit = ems_df2['Unit']
agency_unit = ems_df2['AgencyAndUnit']
```

```python
new_ems_df = pd.DataFrame({"ID": ID,
                           "Hospital_Code": hospital_code,
                          "Date": dates,
                          "Arrival_Date_Time": arrival_time,
                          "APOT": apot,
                          "Impression": impression,
                          "Postal_Code": postal_code,
                          "AgencyNumber": agency,
                          "Agency_Unit": agency_unit,
                          "Lat": lat,
                          "Long": long})
new_ems_df
```

```python
# Identify dtypes
new_ems_df.dtypes
```

```python
new_ems_df.astype({'ID': 'string'}).dtypes
```

```python
new_ems_df = new_ems_df.drop_duplicates(subset=['ID'])
new_ems_df
```

```python
# verify no duplicate rows 
dup_rows = new_ems_df[new_ems_df.duplicated(['ID'])]
dup_rows
```

```python
# Calculate unique values for all columns 
# making sure ID values are all unique 
new_ems_df.nunique()
```
